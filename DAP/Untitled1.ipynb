```python
df = pd.read_csv('D:\DAP\Screentime - Overall Usage.csv')
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,4))

sns.scatterplot(ax=ax[0], data=df, x='Notifications', y='Total Usage')
sns.scatterplot(ax=ax[1], data=df, x='Notifications', y='Unlocks')

plt.show()
```

```python
most_unlocks = df['Unlocks'].max()
print(most_unlocks)

corelated_notifications = df.loc[df['Unlocks'] == most_unlocks, 'Notifications'].iloc[0]
print('No. of times user got notification = ', corelated_notifications, ' Max number of unlocks to their phone =', most_unlocks)
```

```python
sorted_df = df.sort_values(by=['Unlocks'], ascending = True)
sorted_df.head()
```

```python
plt.figure(figsize=(15,8))
plt.style.use("dark_background")
sns.barplot(x='Date ',y='Unlocks', data = sorted_df.head(10))
#sns.set(rc={'figure.figsize':(11.7,8.27)})
plt.yticks(fontsize= 6)
plt.ylabel('Unlocks')
plt.xlabel('Date')
plt.xticks(rotation = 60)
plt.title('Total Unlocks per Day')
```

```python
df = pd.read_csv('D:\DAP\Screentime - App Details.csv')
```

```python
max_times = df['Times opened'].max()
used_app = df.loc[df['Times opened'] == max_times, 'App'].iloc[0]

print(used_app, 'was opened max times ~', max_times, 'times')
```

```python
max_usage = df['Usage'].max()
_app = df.loc[df['Usage'] == max_usage, 'App'].iloc[0]

print(_app, 'had maximum usage ~', max_usage)
```

```python
max_notification = df['Notifications'].max()
u_app = df.loc[df['Notifications'] == max_notification, 'App'].iloc[0]

print(u_app, 'had max notifications ~', max_notification)
```

```python
min_notification = df['Notifications'].min()
us_app = df.loc[df['Notifications'] == min_notification, 'App'].iloc[0]

print(us_app, 'had least notifications ~', min_notification)
```
